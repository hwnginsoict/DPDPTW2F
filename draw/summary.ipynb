```python
import pandas as pd

# Đọc dữ liệu từ CSV
df = pd.read_csv("F:\CodingEnvironment\DPDPTW2F\draw\hv_pivot_last.csv") 

import re

def extract_type_and_node(filename):
    # Sử dụng regex để tách loại (LC, LR, LRC) và node
    # Giả sử: loại luôn bắt đầu bằng "LC" hoặc "LR" hoặc "LRC", theo sau bởi số (nếu có),
    # và sau đó là dấu gạch dưới và số node.
    type_match = re.match(r'^(LRC|LR|LC)', filename)
    node_match = re.search(r'_(\d+)_', filename)
    
    base_type = type_match.group(1) if type_match else None
    node = node_match.group(1) if node_match else None
    
    return base_type, node

df[['base_type', 'node']] = df['file'].apply(lambda x: pd.Series(extract_type_and_node(x)))

# hv_columns = ['MOEA/D', 'MOEA/D+', 'NSGA-II', 'PFG-EMOA']
hv_columns = ['PFG-2F-RK', 'PFG-2F-PM', 'PFG-MOEA','NSGA-II',"MOEA/D","MOEA/D+"]
df[hv_columns] = df[hv_columns].apply(pd.to_numeric, errors='coerce')

# Tính mean và std cho mỗi nhóm base_type và node
agg_df = df.groupby(['base_type', 'node'])[hv_columns].agg(['mean', 'std']).reset_index()
# Gộp các cấp của MultiIndex thành các tên cột duy nhất
agg_df.columns = ['_'.join(col).strip('_') if col[1] else col[0] for col in agg_df.columns.values]

# Tạo cột kết hợp 'type_node'
agg_df['type_node'] = agg_df['base_type'] + '_' + agg_df['node']

# Chọn cột 'type_node' cùng với các cột HV vừa tính toán
selected_columns = ['type_node'] + [col for col in agg_df.columns if col not in ['base_type', 'node', 'type_node']]
final_df = agg_df[selected_columns]

# Đặt 'type_node' làm chỉ mục (index) nếu muốn
final_df.set_index('type_node', inplace=True)

print(final_df)

final_df.to_csv("combined_hv_results.csv")
```
